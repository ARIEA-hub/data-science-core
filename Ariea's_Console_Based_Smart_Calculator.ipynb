# Ariea's Console-Based Smart Calculator

# Problem Statement

The objective of this project is to create a console-based smart calculator using Python and NumPy.

The calculator allows users to perform different mathematical operations such as:
- Addition
- Subtraction
- Multiplication
- Division
- Power
- Modulus

The program is menu-driven, meaning the user can select operations from a menu. The calculator accepts multiple numbers from the user, stores them using lists and NumPy arrays, performs calculations, and displays results.

The project also includes:
- User-defined functions
- Loops
- Conditional statements
- NumPy array operations
- Error handling for division by zero

The calculator continues running until the user chooses to exit.

# Logic Explanation

1. Display a menu of calculator operations.
2. Take the user's choice as input.
3. Ask the user how many numbers they want to enter.
4. Store the numbers in a list.
5. Convert the list into a NumPy array.
6. Use functions to perform calculations.
7. Display the result.
8. Ask the user whether to continue or exit.
9. Repeat the process using a loop until the user exits.

```python
import numpy as np
```

```python
# Function for Addition
def addition(numbers):
    array_numbers = np.array(numbers)
    return np.sum(array_numbers)


# Function for Subtraction
def subtraction(numbers):
    result = numbers[0]

    for i in range(1, len(numbers)):
        result = result - numbers[i]

    return result


# Function for Multiplication
def multiplication(numbers):
    result = 1

    for num in numbers:
        result = result * num

    return result


# Function for Division
def division(numbers):
    result = numbers[0]

    for i in range(1, len(numbers)):
        if numbers[i] == 0:
            return "Error! Division by zero is not allowed."
        result = result / numbers[i]

    return result


# Function for Power
def power(base, exponent):
    return base ** exponent


# Function for Modulus
def modulus(number1, number2):
    return number1 % number2
```

```python
# Main Program

continue_calculator = "yes"

while continue_calculator == "yes":

    print("\n===== SMART CALCULATOR =====")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Power")
    print("6. Modulus")
    print("7. Exit")

    choice = int(input("Enter your choice: "))

    # Addition
    if choice == 1:

        count = int(input("How many numbers do you want to add? "))

        number_list = []

        for i in range(count):
            number = float(input("Enter number: "))
            number_list.append(number)

        result = addition(number_list)

        print("Result =", result)

    # Subtraction
    elif choice == 2:

        count = int(input("How many numbers do you want to subtract? "))

        number_list = []

        for i in range(count):
            number = float(input("Enter number: "))
            number_list.append(number)

        result = subtraction(number_list)

        print("Result =", result)

    # Multiplication
    elif choice == 3:

        count = int(input("How many numbers do you want to multiply? "))

        number_list = []

        for i in range(count):
            number = float(input("Enter number: "))
            number_list.append(number)

        result = multiplication(number_list)

        print("Result =", result)

    # Division
    elif choice == 4:

        count = int(input("How many numbers do you want to divide? "))

        number_list = []

        for i in range(count):
            number = float(input("Enter number: "))
            number_list.append(number)

        result = division(number_list)

        print("Result =", result)

    # Power
    elif choice == 5:

        base = float(input("Enter base number: "))
        exponent = float(input("Enter exponent: "))

        result = power(base, exponent)

        print("Result =", result)

    # Modulus
    elif choice == 6:

        number1 = int(input("Enter first number: "))
        number2 = int(input("Enter second number: "))

        result = modulus(number1, number2)

        print("Result =", result)

    # Exit
    elif choice == 7:

        print("Calculator Closed Successfully!")
        break

    # Invalid Choice
    else:
        print("Invalid Choice! Please try again.")

    continue_calculator = input("Do you want to continue? (yes/no): ")

print("Thank You for Using Smart Calculator")
```
